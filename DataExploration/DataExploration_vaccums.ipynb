```python
# Check the rating graph with the number of reviews
df.rating.value_counts().plot(kind= 'bar', color= 'blue')
figure(figsize= (20,10))
```

```python
df['rating'] = df['rating'].apply(lambda x: x + 1)
```

```python
df
```

```python
df.rating.value_counts().plot(kind= 'bar', color= 'blue')
figure(figsize= (20,10))
```

Mapped the values 0,1,2,3,4 to 1-5 of ratings. As we can see that most of the reviews has 4/5 ratings as compared to 1/2/3

lets consider ratings with 4 and 5 as good and 1,2 and 3 as bad review

```python
df['Review_type']= df.rating
df.Review_type[df.rating <= 3] = 0 #bad review
df.Review_type[df.rating >= 4] = 1 #good review
```

```python
df
```

```python
df.Review_type.value_counts().plot(kind= 'bar', color= 'blue')
figure(figsize= (20,10))
```
