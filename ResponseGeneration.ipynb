```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./TopicModelling/mergedDF_after_TopicModelling.csv')
```

```python
df
```

```python
dict_response = {
    'Positive': ['Thank you so much for your feedback for',
                'Looking forward to hear more about terms ',
                'Hope you enjoyed terms on',
                'I am glad you liked terms '],
    'Negative': ['Apologies for the experience. Sorry about the matters on',
                'Can you please share how we can improve your experience of matters on',
                'Please reach out to complaints for matters on',
                'Can we offer you a replacement? Please contact our customer service for matters on']
}
```

```python
for idx in range(0, len(df['topics'])):
    df['topics'][idx] = df['topics'][idx].split()[:4]
    df['topics'][idx]= ' '.join(df['topics'][idx])
```

```python
from random import randint 
random_index = randint(0,3)

df['response']  = df.reviews
df.response[df.Sentiment == 0] =dict_response['Negative'][random_index] +" "+ df['topics'] + ". Thank you."
df.response[df.Sentiment == 1] = dict_response['Positive'][random_index] +" " + df['topics'] + ". Thank you."
```
